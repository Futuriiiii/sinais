**Exercício 1** Considere o sinal $f(t)$ a tempo contínuo mostrado na Figura 2.6. Apresente um esboço de cada um dos seguintes sinais:



a) $f(t-2)$.

b) $f(3-t)$.

c) $f(3-2 t)$.

```python
# Definindo a função f(t)
def f(t):
    return np.sin(t)  # Exemplo de função

# Criando um vetor de valores de t
t = np.linspace(-2, 5, 400)
```

```python
# Deslocando a função para f(t-2)
f_shifted = f(t - 2)

# Plotando o gráfico
plt.plot(t, f_shifted, label='f(t-2)')
plt.title('Gráfico da função f(t-2)')
plt.xlabel('t [s]')
plt.ylabel('f(t-2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Deslocando a função para f(3-t)
f_shifted = f(3-t)

# Plotando o gráfico
plt.plot(t, f_shifted, label='f(3-t)')
plt.title('Gráfico da função f(3-t)')
plt.xlabel('t [s]')
plt.ylabel('f(3-t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Deslocando a função para f(3-2t)
f_shifted = f(3-2*t)

# Plotando o gráfico
plt.plot(t, f_shifted, label='f(3-2t)')
plt.title('Gráfico da função f(3-2t)')
plt.xlabel('t [s]')
plt.ylabel('f(3-2t)')
plt.legend()
plt.grid(True)
plt.show()
```
