**Exercício 1** Considere o sinal $f(t)$ a tempo contínuo mostrado na Figura 2.6. Apresente um esboço de cada um dos seguintes sinais:



a) $f(t-2)$.

b) $f(3-t)$.

c) $f(3-2 t)$.

```python
def f(t):
    if -2 <= t < -1:
        return 0
    elif -1 <= t < 0:
        return t
    elif 0 <= t < 2:
        return -t + 2
    elif 2 <= t < 4:
        return -1
    elif 4 <= t < 5:
        return 0
    else:
        return np.nan  # Usando NaN para valores fora do intervalo definido

# Gerando valores de t
t_values = np.linspace(-2, 5, 1000)
```

```python
# Definindo a função f(t-2)
def f_shifted(t):
    return f(t - 2)

# Gerando valores de t
t_values_shifted = np.linspace(-2, 5, 1000)
# Calculando os valores de f(t-2) para cada t
f_values_shifted = np.array([f_shifted(t) for t in t_values_shifted])

# Plotando a função f(t-2)
plt.figure(figsize=(10, 6))
plt.plot(t_values_shifted, f_values_shifted, label='f(t-2)')
plt.xlabel('t')
plt.ylabel('f(t-2)')
plt.title('Gráfico da função f(t-2)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Definindo a função f(3-t)
def f_shifted(t):
    return f(3 - t)

# Gerando valores de t
t_values_shifted = np.linspace(-2, 5, 1000)
# Calculando os valores de f(3-t) para cada t
f_values_shifted = np.array([f_shifted(t) for t in t_values_shifted])

# Plotando a função f(3-t)
plt.figure(figsize=(10, 6))
plt.plot(t_values_shifted, f_values_shifted, label='f(3-t)')
plt.xlabel('t')
plt.ylabel('f(3-t)')
plt.title('Gráfico da função f(3-t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Definindo a função f(3-2t)
def f_shifted(t):
    return f(3 - 2*t)

# Gerando valores de t
t_values_shifted = np.linspace(-2, 5, 1000)
# Calculando os valores de f(3-2t) para cada t
f_values_shifted = np.array([f_shifted(t) for t in t_values_shifted])

# Plotando a função f(3-2t)
plt.figure(figsize=(10, 6))
plt.plot(t_values_shifted, f_values_shifted, label='f(3-2t)')
plt.xlabel('t')
plt.ylabel('f(3-2t)')
plt.title('Gráfico da função f(3-2t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# a)
# A função cos(ωt) é periódica se ωt é uma função periódica.
# T = 2π/ω ; ω = sqrt(2)
# T = 2π/sqrt(2) = π * sqrt(2) -> sqrt(2) é irracional
# Logo, T não é períodico
```

```python
# b)
# f(t) será períodica se ambos os termos cos(3t+2) e sin(4t−1) forem periódicos com um período comum
# para cos(3t+2):
# T1 = 2π/3
# para sin(4t−1):
# T2 = 2π/4 = π/2
# Achando o MMC entre T1 e T2:
# MMC(T1,T2) = 6
# Logo, T é periodico com período igual a 6
```

```python
# c)
# Pelo mesmo motivo da letra a) (cos(sqrt(2)t)), essa função não é períodica
```

```python
# d)
# Para pow(e, jωt), o sinal é periódico se ωT=2πk para algum inteiro k
# Para pow(e, j(2π/3)*t):
# ω = 2π/3
# T = 2π/ω = (2π) * (3/2π) = 3
# Logo, T é periodico com período igual a 3
```

```python
# e)
# Para pow(e, (-1 + jω)*t), precisamos que a parte imaginária seja periódica
# para pow(e, (-1+ j2π/3)*t):
# ω = 2π/3
# T = 2π/ω = (2π) * (3/2π) = 3
# Logo, T é periodico com período igual a 3
```
